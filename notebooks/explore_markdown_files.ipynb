```python
import os
```

```python
drugs_to_process = ["CITALOPRAM", "AZITROMICINA"]
```

```python
def read_markdown_file(md_path: str) -> str:
    """
    Reads a Markdown file and returns its content as a string.
    """
    with open(md_path, "r", encoding="utf-8") as file:
        return file.read()
```

```python
import re
from typing import List, Tuple

def extract_flex_numbered_headers(md_text: str) -> List[Tuple[int, str]]:
    """
    Extract headers with section numbers and titles, even if each word is bolded and struck separately:
    Example:
    ~~**1**~~ **.** ~~**DENOMINAZIONE**~~ ~~**DEL**~~ ~~**MEDICINALE**~~
    """
    headers = []

    # Step 1: Normalize strikethrough and bold
    # Convert ~~**WORD**~~ to just WORD
    clean_text = re.sub(r'~~\*\*(.+?)\*\*~~', r'\1', md_text)

    # Step 2: Also remove bold from single elements like **.**
    clean_text = re.sub(r'\*\*(.+?)\*\*', r'\1', clean_text)

    # Step 3: Look for patterns like "1 . TITLE"
    # Match: digit, optional dot, and several words after
    pattern = re.compile(r'^\s*(\d+)\s*[\.\-]?\s+([A-Z][A-Z\s]{3,})$', re.MULTILINE | re.IGNORECASE)

    for match in pattern.finditer(clean_text):
        section = int(match.group(1))
        title = match.group(2).strip().lower()
        headers.append((section, title))

    return headers
```

```python
import pdb 
for drug in drugs_to_process:
    drug_interim_dir = f'../data/interim/{drug}' 

    all_files = os.listdir(drug_interim_dir)
    fi_files = [file for file in all_files if file.startswith('FI_')]

    for markdown_file in fi_files:
        md_text = read_markdown_file(drug_interim_dir + '/' + markdown_file)

        headers = extract_flex_numbered_headers(md_text)

        print(headers)
        print(len(headers))

        if len(headers) > 10:
            print(markdown_file)
        print('-'*100)
```
